```python
import keras 
import numpy as np
import pandas as pd
```

```python
path='C:/Users/Yaswanth/Desktop/american/Problem2' #path for the dataset in the desktop
train_path=path+'/train.csv'
test_path=path+'/test.csv'
```

```python
train_set = pd.read_csv(train_path)
test_set = pd.read_csv(test_path)
y_train_labels = train_set.label.values
train_keys = train_set.key.values
train_set.head()
```

```python
train_set.drop(labels=['label','key'],axis=1,inplace=True)
train_set.head()
```

```python
test_keys = test_set.key.values
test_set.drop(labels=['key'],axis=1,inplace=True)
test_set.head()
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
X_train = standard_scaler.fit_transform(train_set)
X_test = standard_scaler.transform(test_set)
```

```python

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.optimizers import RMSprop,Adam,SGD
```

```python
#adam=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
learning_rate = 0.1
epochs=10
decay_rate = learning_rate / epochs
momentum = 0.8
#sgd = SGD(lr=learning_rate, momentum=momentum, decay=decay_rate, nesterov=False)
adam = Adam(lr=0.001)
model = Sequential()
#model.add(Dropout(0.2, input_shape=(54,)))
model.add(Dense(units=256,activation='relu',input_shape=(54,)))
model.add(Dropout(0.2))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=128,activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=64,activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units = 1,activation="sigmoid"))
model.compile(optimizer =adam,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train_labels,batch_size =25,validation_split=0.2,epochs =10)
```

```python
y_test = model.predict(X_test)
```

```python
y_test
```

```python
acc=model.evaluate(X_test,y_test)
acc
```
